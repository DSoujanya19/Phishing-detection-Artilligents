# Importing libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

# Fetching data and exploring it

```python
data=pd.read_csv("data/Phising_Training_Dataset.csv")
```

```python
data.head()
```

# Splitting inputs and labels/output into diff data frames

```python
labels=data['Result']
labels
```

```python
data.drop(columns=['Result','key'],axis=1,inplace=True)
```

```python
data
```

# splitting train and test data

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
```

```python
trainx,testx,trainy,testy=train_test_split(data,labels,train_size=0.8)
```

# Training the model

```python
model.fit(trainx,trainy)
```

# checking model's performance 

```python
p=model.predict(testx)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(testy,p)
```

```python
c=data.columns
c
```

```python
fig = plt.figure(figsize=(300,110))
_ = tree.plot_tree(model, 
                   feature_names=c,  
                   class_names=['1','-1'],
                   filled=True)
```

# Determining labels for new test data 

```python
testdata=pd.read_csv("data/Phising_Testing_Dataset.csv")
keys = testdata['key']
testdata.drop('key',axis=1,inplace=True)
```

```python
predicted=model.predict(testdata)
```

```python
predicted
```

# Combining key and predicted values and storing in a csv file

```python
testdata['key'] = keys
```

```python
testdata['Result'] = predicted
```
